```python
# Glob is library to quickly get a list of filenames
import glob

# Regular expressions
import re

# Nibabel can load .nii.gz-files ('nifti files')
import nibabel as nb

# Pandas is a data analysis
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Get a list fo possible filenames
#fns = glob.glob('/path/to/contrasts/ds*/contrasts*/threshold_*/zstats_1.nii.gz')
fns = glob.glob('/home/felice/level2_fsl_results2/ds*/contrast*/thresholded_z/_threshold_*/zstat1_threshold.nii.gz')


# Put values in lists
ds = []
contrasts = []
thresholds = []
for fn in fns:
    d_value = re.match('.*/ds(?P<ds>[0-9]+)', fn)
    c_value = re.match('.*/contrast(?P<contrast>[0-9]+)', fn)
    t_value = re.match('.*/_threshold_(?P<threshold>[0-9]+\.[0-9])', fn) 
    
    ds.append(d_value.groupdict()['ds'])
    contrasts.append(c_value.groupdict()['contrast'])
    thresholds.append(t_value.groupdict()['threshold'])
    
# Create empty list that we are gonna fill with dictionaries with data
# We will convert it to a pandas dataframe later
results = []

mni_brain = nb.load('/usr/share/fsl/data/standard/MNI152_T1_2mm_brain_mask.nii.gz').get_data()
vol_brain = (mni_brain > 0).sum() / 0.8
# iterate over values and filenames
for fn, d, contrast, thr in zip(fns, ds, contrasts, thresholds):
 
     # load z-map
     z_map = nb.load(fn).get_data()

     # Construct corresponding in limbo filename and load z-map
     in_limbo = nb.load('/home/felice/level2_fsl_results2/ds%s/contrast%s/in_limbo/_threshold_%s/in_limbo.nii.gz' % (d, contrast, thr)).get_data()

     # Count number of boxels above thresholds
     n_above_thresholds = (z_map > 0).sum()
        
     # Count % of brain above thresholds
     per_above_thresholds = (n_above_thresholds * 100)/vol_brain

     # Count number of in limbo voxels
     n_in_limbo = (in_limbo > 0).sum()
        
     #Count % brain in limbo zone
     per_in_limbo = (n_in_limbo * 100)/vol_brain

     # Put results in list of dictionaries
     results.append({'ds':d, 'contrast':contrast, 'n_voxel': n_above_thresholds, 'volume %': per_above_thresholds, 'threshold':thr, 'categories':'above_thresholds'})
     results.append({'ds':d, 'contrast':contrast, 'n_voxel': n_in_limbo, 'volume %': per_in_limbo, 'threshold':thr, 'categories':'in_limbo'})
     #results.append({'ds':d, 'contrast':contrast, 'z_above_thr': n_above_thresholds, 'threshold':thr, 'n_in_limbo':n_in_limbo})  

        
# Create pandas dataframe
df = pd.DataFrame(results)
    
```

```python
def plot_brain(which_slice, slice_axis='y', data=mni_brain, **kwargs):
    
    # (**kwargs is a dictionary of all other keywords that given to the function
    # we can later push these through to imshow by using same syntax
    
    # We want to set cmap to gray if not explicity passed to the function
    if 'cmap' not in kwargs.keys():
        kwargs['cmap'] = plt.cm.gray
    
    # Interpolation to nearest neighbor gives nicer plot
    if 'interpolation' not in kwargs.keys():
        kwargs['interpolation'] = 'nearest'
    
    index_tuple = [slice(None), slice(None), slice(None)]
    
    slice_axis = {'x':0, 'y':1, 'z':2}[slice_axis]
    
    index_tuple[slice_axis] = which_slice
    

   
    plt.imshow(data[index_tuple].T, origin='lower', **kwargs)
    
```
